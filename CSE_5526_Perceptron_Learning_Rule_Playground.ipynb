```
import numpy as np
import matplotlib.pyplot as plt

x=np.array([[1,3],[-1,5],[-3,2],[5,1],[4,-2],[1,-3]])
d=np.array([1,1,1,-1,-1,-1])
w=np.array([8,-2])

def plotdata(xd,dd,wd):

  fig,ax = plt.subplots(figsize=[6,6])
  ax.grid(color=[.7,.7,.7],linestyle='-')
  r=10
  ax.set_xlim([-r,r])
  ax.set_ylim([-r,r])
  ax.set_xticks(np.arange(-r,r+1,1))
  ax.set_yticks(np.arange(-r,r+1,1))
  ax.plot(x[d==1,0],x[d==1,1],'rx',x[d==-1,0],x[d==-1,1],'bo',
          [-r,r],[0,0],'k-',[0,0],[-r,r],'k-')
  ax.annotate("", xy=(wd[0], wd[1]), xytext=(0, 0), arrowprops=dict(arrowstyle="->"))
  xv=np.array([-100,100])
  yv=-wd[0]/wd[1]*xv
  ax.plot(xv,yv,'k--')
  return fig,ax

f,a=plotdata(x,d,w)
```
